```python
import base64
import io
import gym
from gym.wrappers.monitoring.video_recorder import VideoRecorder
from IPython import display
import time
import os
```

```python
now_time = time.time()
save_dir = f"./{now_time}"
print(save_dir)
save_path = save_dir + "/0.mp4"
os.makedirs(save_dir, exist_ok=True)

env = gym.make('CartPole-v1')
recorder = VideoRecorder(env, save_path, enabled=True)

o = env.reset()
for _ in range(1000):
    env.render()
    recorder.capture_frame()
    o, r, d, i = env.step(env.action_space.sample()) # 本当はDNNからアクションを入れる
    if d:
        env.reset()

recorder.enable = False
recorder.close()
env.close()
```

```python
mp4_files = [file for file in os.listdir(save_dir) if file.endswith(".mp4")]
mp4_files
```

```python
print(mp4_files)
for f in mp4_files:
    f = save_dir + "/" + f
    print(f)
    video = io.open(f, 'r+b').read()
    encoded = base64.b64encode(video)

    display.display(display.HTML(data="""
        <video alt="test" controls>
        <source src="data:video/mp4;base64,{0}" type="video/mp4" />
        </video>
        """.format(encoded.decode('ascii'))))
```
